<h1 style="font-family: Trebuchet MS; padding: 12px; font-size: 32px; color: #FF5A5F; line-height: 1.25;"><b>Understanding Given Variables :</b></h1>

<span style="font-family: Trebuchet MS; font-size: 18px; color: #FF5A5F; line-height: 1.25;"><b>id:</b> A unique identifier assigned to each listing in the dataset.</span>

<span style="font-family: Trebuchet MS; font-size: 18px; color: #FF5A5F; line-height: 1.25;"><b>name:</b> The title of the listing, as it appears on Airbnb.</span>

<span style="font-family: Trebuchet MS; font-size: 18px; color: #FF5A5F; line-height: 1.25;"><b>host_id:</b> A unique identifier assigned to each host in the dataset.</span>

<span style="font-family: Trebuchet MS; font-size: 18px; color: #FF5A5F; line-height: 1.25;"><b>host_name:</b> The name of the host, as displayed on Airbnb.</span>

<span style="font-family: Trebuchet MS; font-size: 18px; color: #FF5A5F; line-height: 1.25;"><b>neighbourhood_group:</b> The broader neighborhood area within New York City, such as Manhattan or Brooklyn.</span>

<span style="font-family: Trebuchet MS; font-size: 18px; color: #FF5A5F; line-height: 1.25;"><b>neighbourhood:</b> The specific neighborhood location of the listing.</span>

<span style="font-family: Trebuchet MS; font-size: 18px; color: #FF5A5F; line-height: 1.25;"><b>latitude:</b> The geographical latitude coordinate of the listing.</span>

<span style="font-family: Trebuchet MS; font-size: 18px; color: #FF5A5F; line-height: 1.25;"><b>longitude:</b> The geographical longitude coordinate of the listing.</span>

<span style="font-family: Trebuchet MS; font-size: 18px; color: #FF5A5F; line-height: 1.25;"><b>room_type:</b> The type of accommodation offered, such as an entire home, private room, or shared room.</span>

<span style="font-family: Trebuchet MS; font-size: 18px; color: #FF5A5F; line-height: 1.25;"><b>price:</b> The nightly rental price for the listing, in US dollars.</span>

<span style="font-family: Trebuchet MS; font-size: 18px; color: #FF5A5F; line-height: 1.25;"><b>minimum_nights:</b> The minimum number of nights a guest is required to book.</span>

<span style="font-family: Trebuchet MS; font-size: 18px; color: #FF5A5F; line-height: 1.25;"><b>number_of_reviews:</b> The total number of reviews the listing has received.</span>

<span style="font-family: Trebuchet MS; font-size: 18px; color: #FF5A5F; line-height: 1.25;"><b>reviews_per_month:</b> The average number of reviews the listing receives each month.</span>

<span style="font-family: Trebuchet MS; font-size: 18px; color: #FF5A5F; line-height: 1.25;"><b>calculated_host_listings_count:</b> The total number of listings that a host has on Airbnb.</span>

<span style="font-family: Trebuchet MS; font-size: 18px; color: #FF5A5F; line-height: 1.25;"><b>availability_365:</b> The number of days the listing is available for booking in the upcoming year (365 days).</span>

<span style="font-family: Trebuchet MS; font-size: 24px; color: #FF5A5F; line-height: 1.25;">Checking NULL values in Dataset</span>

```python
airbnb_df.isnull().sum()
```

<span style="font-family: Trebuchet MS; font-size: 20px; color: #FF5A5F; line-height: 1.25;"><b>NOTE :</b></span>
<span style="font-family: Trebuchet MS; font-size: 20px; color: #FF5A5F; line-height: 1.25;"><b>host_name</b> and <b>listing_name</b> have not large amount of NULL values, so firstly fill those with some substitutes in both the columns.</span>

```python
airbnb_df['listing_name'].fillna('unknown',inplace=True)
airbnb_df['host_name'].fillna('no_name',inplace=True)
```

```python
airbnb_df[['host_name','listing_name']].isnull().sum()
```

<span style="font-family: Trebuchet MS; font-size: 20px; color: #FF5A5F; line-height: 1.25;"><b>NOTE :</b></span>
<span style="font-family: Trebuchet MS; font-size: 20px; color: #FF5A5F; line-height: 1.25;">Now, the columns <b>last_review</b> and <b>reviews_per_month</b> have total 10052 null values each.</span>
<span style="font-family: Trebuchet MS; font-size: 20px; color: #FF5A5F; line-height: 1.25;"><b>last_review</b> column is not required for our analysis as compared to <b>number_of_reviews</b> & <b>reviews_per_month</b>. We're good to drop this column.</span>

```python
airbnb_df = airbnb_df.drop(['last_review'], axis=1)
```

```python
airbnb_df.info()
```

<span style="font-family: Trebuchet MS; font-size: 20px; color: #FF5A5F; line-height: 1.25;"><b>NOTE :</b></span>
<span style="font-family: Trebuchet MS; font-size: 20px; color: #FF5A5F; line-height: 1.25;">The <b>reviews_per_month</b> column also containing NULL/NaN values and we can simple put 0 reviews by replacing NAN's</span>

```python
airbnb_df['reviews_per_month'] = airbnb_df['reviews_per_month'].replace(to_replace=np.nan,value=0).astype('int64')
```

```python
airbnb_df['reviews_per_month'].isnull().sum() 
```

<span style="font-family: Trebuchet MS; font-size: 20px; color: #FF5A5F; line-height: 1.25;"><b>NOTE :</b></span>
<span style="font-family: Trebuchet MS; font-size: 20px; color: #FF5A5F; line-height: 1.25;">So, there is no null value now in <b>reviews_per_month</b> column because we replaced NULL/NaN value by 0.</span>
